```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
def show_depth_map(leftImage, rightImage, numDisparities, blockSize):
    imgL = cv2.imread(leftImage,0)
    imgR = cv2.imread(rightImage,0)
    stereo = cv2.StereoBM_create(numDisparities, blockSize)
    disparity = stereo.compute(imgL,imgR)
    plt.imshow(disparity,'gray')
    plt.title("Disparities = " + str(numDisparities) + " & Block Size = " + str(blockSize)) 
    # plt.show()
```

```
show_depth_map("stereo_images/1left.png", "stereo_images/1right.png", 16, 15)
```
